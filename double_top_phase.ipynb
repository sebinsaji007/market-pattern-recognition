```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
crypto_symbol = 'BTC-USD'
crypto_data = yf.download("BTC-USD", start="2023-07-20", end="2023-07-22",  interval = "60m")  # Fetch one year of historical data
```

```python
crypto_data['whole_pricechange'] = crypto_data['Close'] - crypto_data['Close'].shift(1)
crypto_data
```

```python
def find_reversal_points(data):
    reversal_points = []

    for i in range(1, len(data)):
        if (data['PriceChange'][i] >= 0 and data['PriceChange'][i - 1] < 0) or (data['PriceChange'][i] <= 0 and data['PriceChange'][i - 1] > 0):
            reversal_points.append(data.index[i])


    return reversal_points
```

```python
reversal_points = find_reversal_points(crypto_data)

plt.figure(figsize=(12, 6))
plt.plot(crypto_data.index, crypto_data['Close'], label='Close Price', color='b')
plt.scatter(reversal_points, crypto_data.loc[reversal_points]['Close'], color='r', label='Reversal Points')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Price')
plt.title(f'{crypto_symbol} Price with Reversal Points')
plt.grid()
plt.show()
```

```python
crypto_data['pricechange'] = crypto_data.loc[reversal_points, 'Close']-crypto_data.loc[reversal_points, 'Close'].shift(1)
crypto_data.loc[reversal_points, ['Close','pricechange']]
```

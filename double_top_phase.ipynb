```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import pytz
```

```python
crypto_symbol = 'BTC-USD'
crypto_data = yf.download("BTC-USD", start="2023-07-20", end="2023-07-22",  interval = "60m")  # Fetch one year of historical data
```

```python
tz = pytz.timezone('Asia/Kolkata')
crypto_data.index = crypto_data.index.tz_localize(pytz.utc).tz_convert(tz)
```

```python
crypto_data['whole_pricechange'] = crypto_data['Close'] - crypto_data['Close'].shift(1)
crypto_data
```

```python
crypto_data['PriceChange'] = crypto_data['Close'].pct_change()
```

```python
def find_reversal_points(data):
    reversal_points = []

    for i in range(1, len(data)):
        if (data['PriceChange'][i] >= 0 and data['PriceChange'][i - 1] < 0) or (data['PriceChange'][i] <= 0 and data['PriceChange'][i - 1] > 0):
            reversal_points.append(data.index[i])


    return reversal_points
```

```python
reversal_points = find_reversal_points(crypto_data)

plt.figure(figsize=(12, 6))
plt.plot(crypto_data.index, crypto_data['Close'], label='Close Price', color='b')
plt.scatter(reversal_points, crypto_data.loc[reversal_points]['Close'], color='r', label='Reversal Points')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Price')
plt.title(f'{crypto_symbol} Price with Reversal Points')
plt.grid()
plt.show()
```

```python
crypto_data['pricechange_rev'] = crypto_data.loc[reversal_points, 'Close']-crypto_data.loc[reversal_points, 'Close'].shift(1)
crypto_data.loc[reversal_points, ['Close','pricechange_rev']]
```

```python
result_array = []

for t1_index, t1 in enumerate(reversal_points[:-1]):
    for t2 in reversal_points[t1_index + 1:]:
        t1_close = crypto_data.loc[t1, 'Close']
        t2_close = crypto_data.loc[t2, 'Close']
        lowest_close = crypto_data.loc[t1:t2, 'Close'].min()

        if t1_close <= lowest_close or t2_close <= lowest_close or (t1_close < 1.003 * lowest_close and t2_close < 1.003 * lowest_close):
            continue

        # Check if any Close price from t2 onwards breaks the lowest_close
        break_found = False
        for close_price in crypto_data.loc[t2:, 'Close']:
            if close_price < lowest_close:
                print(f"Index: {index}, t1: {t1}, t2: {t2}, Close(t1): {t1_close}, Close(t2): {t2_close}, Lowest Close: {lowest_close}")
                print("---------------------------------------------------------------------------------------------------------------")
                break_found = True
                break

        if break_found:
            continue

        result_array.append([t1, t2, t1_close, t2_close, lowest_close])

# Print the result array with index
#for index, (t1, t2, t1_close, t2_close, lowest_close) in enumerate(result_array, 1):
#  print(f"Index: {index}, t1: {t1}, t2: {t2}, Close(t1): {t1_close}, Close(t2): {t2_close}, Lowest Close: {lowest_close}")
```

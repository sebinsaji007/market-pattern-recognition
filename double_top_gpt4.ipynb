```python
def streamlined_double_top(data, threshold=0.01, window_size=100):
    double_top_patterns = []
    n = len(data)

    # Identify all peaks
    peaks = [i for i in range(1, n-1) if data['High'][i] > data['High'][i-1] and data['High'][i] > data['High'][i+1]]

    # Check pairs of peaks for double top pattern within the window
    for i in range(len(peaks)):
        for j in range(i+1, min(i+window_size, len(peaks))):
            # Check if peaks are close enough in value
            if abs(data['High'][peaks[i]] - data['High'][peaks[j]]) / data['High'][peaks[i]] <= threshold:
                trough_index = peaks[i] + (peaks[j] - peaks[i]) // 2
                # Check if the trough value is lower than the peaks
                if data['Low'][trough_index] < data['High'][peaks[i]] and data['Low'][trough_index] < data['High'][peaks[j]]:
                    # Check for confirmation after second peak
                    post_peak_values = data['Low'][peaks[j]:min(peaks[j] + window_size, n)]
                    if post_peak_values.min() < data['Low'][trough_index]:
                        double_top_patterns.append((peaks[i], trough_index, peaks[j]))

    return double_top_patterns
```

```python
# Find double top patterns using the streamlined function
streamlined_patterns = streamlined_double_top(data)

# Extract the relevant data points for the identified patterns
streamlined_data = []
for pattern in streamlined_patterns:
    streamlined_data.append(data.iloc[list(pattern)])

streamlined_data
```
